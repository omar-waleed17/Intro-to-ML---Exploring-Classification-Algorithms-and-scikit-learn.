Decision tree classifier
adv 1-easy to interpret and visualise 
    2-can handle non linear relations
disadv 1-: Decision trees can easily overfit noisy data, resulting in poor generalization.
       2-Decision trees are prone to overfitting, especially when they are deep and complex. This can lead to poor generalization to unseen data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_decision_regions

# Function to plot the decision boundary
def plot(X, y, model, feature1, feature2, title):
    plot_decision_regions(X, y, clf=model, legend=0)
    plt.xlabel(feature1)
    plt.ylabel(feature2)
    plt.title(title)
    plt.show()

# 1. Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Use only the first two features (sepal length and sepal width)
y = iris.target  # Labels (species)

# 2. Train/Test Split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=True)

# 3. Train the Decision Tree classifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

# 4. Use the plot function to visualize the decision boundary
plot(X_train, y_train, clf, "Sepal length", "Sepal width", "Decision Tree Decision Boundary")

# 5. Predict labels for the test set
y_pred = clf.predict(X_test)

# 6. Evaluate the model accuracy on the test set
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {accuracy * 100:.2f}%")


```
