adv of knn 1-easy to understand and implement
           2- KNN is non-parametric, meaning it does not make any assumptions about the underlying data distribution. This makes it suitable for a wide range of problems.
           3-KNN can be used for both classification and regression tasks.


disadv of knn 1-As the size of the dataset grows, the prediction time can increase significantly since KNN needs to compute the distance   between  the query point and all other points in the training set.
              2-KNN can perform poorly if there are irrelevant or redundant features in the dataset, as these can affect the distance calculation.
              3- stores all training data, making it memory-intensive for large datasets.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_decision_regions

# Function to plot the decision boundary
def plot(X, y, model, feature1, feature2, title):
    plot_decision_regions(X, y, clf=model, legend=0)
    plt.xlabel(feature1)
    plt.ylabel(feature2)
    plt.title(title)
    plt.show()

# 1. Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Use only the first two features (sepal length and sepal width)
y = iris.target  # Labels (species)

# 2. Train/Test Split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=True)

# 3. Train the KNN classifier (k=9)
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train, y_train)

# 4. Use the plot function to visualize the decision boundary
plot(X_train, y_train, knn, "Sepal length", "Sepal width", "KNN Decision Boundary (k=9)")

# 5. Predict labels for the test set
y_pred = knn.predict(X_test)

# 6. Evaluate the model accuracy on the test set using knn.score()
accuracy_score_builtin = knn.score(X_test, y_test)
print(f"Accuracy on test set using knn.score(): {accuracy_score_builtin * 100:.2f}%")

# 7. Evaluate the model accuracy manually using accuracy_score
accuracy_score_manual = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set using accuracy_score: {accuracy_score_manual * 100:.2f}%")

```
